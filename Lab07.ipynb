```python
#W7A1
def binary_search(arr, l, r, target):
    mid = 0
    while l <= r:
        mid = (l + r) // 2
        if target == arr[mid]:
            return mid
        elif arr[mid] < target: l = mid + 1
        else: r = mid - 1
    
    return -1

arr = list(map(int, input().split()))
n = int(input())
print(binary_search(arr, 0, len(arr) - 1, n))
```

```python
#W7A2
def count_occurrences(arr, x):
    cnt = 0
    for i in arr:
        cnt += i == x
    return cnt

arr = list(map(int, input().split()))
x = int(input())
print(count_occurrences(arr, x))
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped: break

def selection_sort(arr):
    n = len(arr)
    for i in range(0, n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    
li = list(map(int, input().split()))
selection_sort(li)
print(li)
```

```python
#W7A8
def bin_se(x, pa):
    le = 0
    ri = len(pa) - 1
    
    ans = -1
    while le <= ri:
        mid = (le + ri) // 2
        if x <= pa[mid][1]:
            if x >= pa[mid][0]:
                ans = pa[mid][1] - pa[mid][0] + 1
                ri = mid - 1
            else: le = mid + 1
        else:
            le = mid + 1
    return ans


pa = eval(input())
pa.sort(key = lambda x: (x[1], x[1] - x[0] + 1))
li = eval(input())
for x in li:
    print(bin_se(x, pa), end = " ")
```

```python
#W7A9
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.smaller = 0
        self.dup = 1
    
    def xuli(self, node, val):
        cnt = 0
        while True:
            if val < node.val:
                node.smaller += 1
                if node.left == None:
                    node.left = Node(val)
                    break
                else:
                    self.xuli(node.left, val)
            elif val > node.val:
                cnt += node.smaller + node.dup
                if node.right == None:
                    node.right = Node(val)
                else:
                    self.xuli(node.right, val)
            else: 
                node.dup += 1
                cnt += node.smaller
        return cnt

def xuli(node, val):
    cnt = 0
    while True:
        if val < node.val:
            node.smaller += 1
            if node.left == None:
                node.left = Node(val)
                break
            else:
                node = node.left
        elif val > node.val:
            cnt += node.smaller + node.dup
            if node.right == None:
                node.right = Node(val)
                break
            else:
                node = node.right
        else:
            node.val += 1
            cnt += node.smaller
            break
    return cnt

def solve(li):
    ans = []
    root = Node(li[len(li) - 1])
    ans.append(0)
    for i in range(len(li) - 2, -1, -1):
        ans.append(xuli(root, li[i]))
    return ans


li = list(map(int, input().split()))
ans = solve(li)
for i in range(len(ans) - 1, -1, -1):
    print(ans[i], end = " ")
```

```python
#W7A10
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def xuli(node, val, m):
    ans = 0
    while True:
        if val < node.val:
            ans = max(ans, m - (node.val - val))
            if node.left == None:
                node.left = Node(val)
                break
            else:
                node = node.left
        else:
            ans = max(ans, val)
            if node.right == None:
                node.right = Node(val)
                break
            else: 
                node = node.right
        
    return ans

li = list(map(int, input().split()))
m = int(input())
s = li[0] % m
root = Node(s)
ans = 0
for i in range(len(li)):
    s += li[i]
    s %= m
    ans = max(ans, xuli(root, s, m))
print(ans)
```
