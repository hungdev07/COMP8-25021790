```python
def Max_Of_Two(a, b):
    if a > b: return a
    return b
```

```python
def PositiveNumbers():
    assert Max_Of_Two(34, 56) == 56, "The max of 34 and 56 should be 56"
    assert Max_Of_Two(234, 345) == 345, "The max of 234 and 345 should be 345"

def NegativeNumbers():
    assert Max_Of_Two(-11, -34) == -11, "The max of -11 and -34 should be -11"
    assert Max_Of_Two(-90, -999) == -90, "The max of -90 and -999 should be -90"
def PosAndNega():
    assert Max_Of_Two(34, -90) == 34, "The max of 34 and -90 should be 34"
    assert Max_Of_Two(0, -9) == 0, "The max of 0 and -9 should be 0"

if __name__ == "__main__":
    PositiveNumbers()
    NegativeNumbers()
    PosAndNega()
    print("pass")
```

```python
def check(n):
    s = 0
    for i in range(1, n):
        if n % i == 0:
            s += i
    return s == n
```

```python
def sohoanhao():
    assert check(23) == False, "check(23) should be False"
    assert check(6) == True, "check(6) should be False"
    assert check(123) == False, "check(123) should be False"

if __name__ == "__main__":
    sohoanhao()
    print("pass")
```

W5A6

```python
mp = {}
def factorial(n):
    global mp
    if n == 0: return 1
    if n in mp: return mp[n]
    mp[n] = n * factorial(n - 1)
    return mp[n]
```

```python
def first():
    assert factorial(0) == 1, "Factorial(0) should be 1"

def small_numbers():
    assert factorial(2) == 2, "Factorial(2) should be 2"
    assert factorial(3) == 6, "Factorial(3) should be 6"

def big_numbers():
    assert factorial(5) == 120, "Factorial(5) should be 120"
    assert factorial(10) == 3628800, "Factorial(10) should be 3628800"

if __name__ == "__main__":
    first()
    small_numbers()
    big_numbers()
    print("pass")
```

```python
def ktra_dau(c):
    if c == "+": return 1
    elif c== "-": return 2
    elif c == "*": return 3
    return 4


def calculator(s: str):
    st = []
    idx = 0
    for i in range(len(s)):
        if s[i] == " ":
            idx = i
            break 

    chuoi = s[:idx]
    st.append(float(chuoi))
    idx += 1
    dau = ktra_dau(s[idx])
    chuoi = ""
    for i in range(idx, len(s)):
        if s[i] == " ":
            if chuoi != "":
                if dau == 1:
                    st.append(float(chuoi))
                elif dau == 2:
                    st.append(-float(chuoi))
                elif dau == 3:
                    a = st[len(st) - 1]
                    st.pop()
                    b = float(chuoi)
                    a = a * b
                    st.append(a)
                else:
                    a = st[len(st) - 1]
                    st.pop()
                    b = float(chuoi)
                    a = a / b
                    st.append(a)
                chuoi = ""
            else: continue
        elif (s[i] >= '0' and s[i] <= '9') or s[i] == ".":
            chuoi += s[i]
        else: 
            if s[i + 1] >= "0" and s[i + 1] <= '9': chuoi += s[i]
            else: dau = ktra_dau(s[i])
    if dau == 1:
        st.append(float(chuoi))
    elif dau == 2:
        st.append(-float(chuoi))
    elif dau == 3:
        a = st[len(st) - 1]
        st.pop()
        b = float(chuoi)
        a = a * b
        st.append(a)
    else:
        a = st[len(st) - 1]
        st.pop()
        b = float(chuoi)
        a = a / b
        st.append(a)
    
    ans = 0
    for i in st:
        ans += i
    
    return float("{:.2f}".format(ans))

print(calculator("12 / 4"))


```

```python
from math import isclose
def check():
    assert calculator("2 + 3") == 5.00, "Test1 Failed"
    assert calculator("12 / 4") == 3.00, "Test2 Failed"
    assert calculator("23.89 * 23.12") == 552.34, "Test3 Failed"
    assert calculator("82 - 12") == 70.00, "Test4 Failed"
    assert calculator("12 * -23.004") == -276.05, "Test5 Failed"

if __name__ == "__main__":
    check()
    print("pass")
```

```python
def Hamming(x, y):
    x = x ^ y
    ans = 0
    while x > 0:
        ans += x % 2
        x = int(x / 2)
    return ans
```

```python
def check():
    assert Hamming(1, 4) == 2, "Test1 Failed"
    assert Hamming(123456, 32321) == 6, "Test2 Failed"
    assert Hamming(1111, 1111) == 0, "Test3 Failed"
    assert Hamming(87654321, 21321835) == 15, "Test4 Failed"
    assert Hamming(75849321, 45687214) == 18, "Test5 Failed"

if __name__ == "__main__":
    check()
    print("pass")
```

```python
def tong(n):
    s = str(n)
    ans = 0
    for i in s:
        ans += int(i)
    return ans
```

```python
def check():
    assert tong(1234) == 10, "Test1 failed"
    assert tong(10000) == 1, "Test2 failed"
    assert tong(98765) == 35, "Test3 failed"
    assert tong(0o1) == 1, "Test4 failed"
    assert tong(10045995) == 33, "Test5 failed"

if __name__ == "__main__":
    check()
    print("pass")
```

```python
def dangcau(s1, s2):
    mp1 = {}
    mp2 = {}
    for i in range(len(s1)):
        if s1[i] in mp1:
            if mp1[s1[i]] != s2[i]: return "false"
        else: mp1[s1[i]] = s2[i]
        if s2[i] in mp2:
            if mp2[s2[i]] != s1[i]: return "false"
        else: mp2[s2[i]] = s1[i]
    return "true"

print(dangcau("egg", "add"))
```

```python
def check():
    assert dangcau("egg", "add") == "true", "Test1 failed"
    assert dangcau("paper", "title") == "true", "Test2 failed"
    assert dangcau("foo", "bar") == "false", "Test3 failed"
    assert dangcau("show", "same") == "true", "Test4 failed"
    assert dangcau("show", "seem") == "false", "Test5 failed"

if __name__ == "__main__":
    check()
    print("pass")
```
