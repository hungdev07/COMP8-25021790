```python
import json

d1 = input()
d2 = input()

mp1 = eval(d1)
mp2 = eval(d2)

mp = {}
for key in mp1.keys():
    if key in mp2:
        mp[key] = mp1[key] + mp2[key]
    else: mp[key] = mp1[key]
for key in mp2.keys():
    if key in mp1:
        mp[key] = mp1[key] + mp2[key]
    else: mp[key] = mp2[key]

tmp = sorted(mp.keys())
ans = {key: mp[key] for key in tmp}
print(ans)
```

```python
#W6A10
arr = list(map(int, input().split()))
k = int(input())
li = []
mp = {}
for i in arr:
    if i not in mp:
        mp[i] = 1
        li.append(i)
    else: mp[i] += 1
ans = []
for i in li:
    if k - i in mp:
        if k - i == i and mp[i] > 1:
            ans.append((i, i))
            break
        elif k - i > i:
            ans.append((i, k - i))
            del mp[i]
print(ans)
```

```python
#W6A12
arr = map(int, input().split())
mp = {}
ans = -1
idx = -1
for i in arr:
    mp[i] = mp.get(i, 0) + 1
    if mp[i] >= ans:
        if mp[i] == ans:
            if i < idx:
                idx = i
        else:
            idx = i
            ans = mp[i]
print(idx)
```

```python
#W6A17
n = int(input())
m = n
li = []
for i in range(n):
    tmp = list(map(int, input().split()))
    li.append(tmp)
for i in range(n):
    for j in range(m):
        if i == j:
            print(li[i][j], end = " ")
print(end = "\n")
for i in range(n):
    for j in range(m):
        if i + j == n - 1:
            print(li[i][j], end = " ")
```

```python
#W6A18
n, m, k = map(int, input().split())
li = []
for i in range(n):
    tmp = list(map(int, input().split()))
    li.append(tmp)
cot = 0
for i in range(n):
    cot += li[i][k]
hang = 0
for i in range(m):
    hang += li[k][i]
print(f"Sum of col k: {cot}")
print(f"Sum of row k: {hang}")
```

```python
#W6A20
def solve(li, k):
    for i in range(len(li)):
        if li[i] == k:
            return i + 1
    return -1
arr = list(map(int, input().split()))
k = int(input())
print(solve(arr, k))
```
