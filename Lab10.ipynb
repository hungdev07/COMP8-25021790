```python
#W10A3

def calc(a):
    s = str(a)
    ans = 0
    for i in s:
        ans += int(i) * int(i)
    return ans

def check(a):
    se = set()
    se.add(a)
    while a != 1:
        a = calc(a)
        if a in se:
            raise ValueError
        se.add(a)
    return True

a = int(input())
try:
    if check(a):
        print("YES")
except ValueError:
    print("NO")
```

```python
#W10A4
def check(a, b):
    if a.isdigit() and not b.isdigit():
        raise ValueError("b is not digit")
    elif not a.isdigit() and b.isdigit():
        raise ValueError("a is not digit")
    elif not a.isdigit() and not b.isdigit():
        raise ValueError("a and b are not digit")
    print(int(a) + int(b))

a, b = input().split()
try:
    check(a, b)
except ValueError as e:
    print(e)
finally:
    print("Kết thúc chương trình")
```

```python
#W10A10
def calc(se):
    ans = 0
    for i in se:
        ans += i
    return ans

li = list(map(int, input().split()))
se = set()
for i in range(1, len(li)):
    se.add(li[i])
if len(se) == li[0]:
    print(calc(se))
else:
    print("Mang khong hop le")
```

```python
#W10A11
'''
*: Bộ định lượng, sau dấu này thì tần suất của kí tự trước dấu xuất hiện từ 0, 1, 2, 3
+: Bộ định lượng, như dấu * nhưng tần suất phải từ 1, 2, 3, 4
.: Là một kí tự đại diện, nó có thể đại diện cho bất kỳ ký tự nào
\\: Ký tự thoát, có nhiệm vụ "vô hiệu hoá" ý nghĩa đặc biệt của kí tự sau nos
'''
def solve(s):
    ok = 1

    if s.endswith("\\") and not s.endswith("\\\\"):
       return False
    
    st = []
    can_repeat = False
    i = 0
    while i < len(s):
        c = s[i]
        if c == "\\":
            i += 1
            can_repeat = True
        elif c in "([":
            can_repeat = False
            st.append(c)
        elif c == ')':
            if len(st) == 0:
                return False
            if s[-1] != '(': return False
            st.pop()
            can_repeat = True
        elif c == ']':
            if len(st) == 0:
                return False
            if s[-1] != '[': return False
            st.pop()
            can_repeat = True
        elif c in "?+*":
            if not can_repeat:
                return False
            else:
                can_repeat = False
        else:
            can_repeat = True
        i += 1
    return len(st) == 0


for _ in range(int(input())):
    s = input()
    if solve(s):
        print("Valid")
    else:
        print("Invalid")
```
