```python
#W8A1
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14
    def dttp (self):
        return 2 * self.pi * self.radius * (self.radius + self.height)
    def volume (self):
        return self.pi * (self.radius ** 2) * self.height

r, h = map(int, input().split())
c = Cylinder(r, h)
print(f"{c.dttp():.2f}, {c.volume():.2f}")
```

```python
#W8A2
def nhuan(year):
    if year % 4 == 0:
        if year % 400 == 0: return True
        elif year % 100 == 0: return False
        return True

n31 = [1, 3, 5, 7, 8, 10, 12]
n30 = [4, 6, 9, 11]

class Date:
    def __init__ (self, date):
        self.day = int(date[:2])
        self.month = int(date[3:5])
        self.year = int(date[6:])
    
    def valid (self):
        if self.month > 12: return False
        elif self.month == 2:
            if nhuan(self.year):
                if self.day == 29: return True
                else: return self.day == 28
        elif self.month in n31:
            return self.day == 31
        elif self.month in n30:
            return self.day == 30
    
    def ngayketiep(self):
        if self.day == 31 and self.month == 12:
            s = str(self.year + 1)
            while len(s) < 4:
                s = "0" + s
            return f"01/01/{s}"
        elif (self.month in n31 and self.day == 31) or (self.month in n30 and self.day == 30): 
            if self.month < 9: return f"01/0{self.month + 1}/{self.year}"
            return f"01/{self.month + 1}/{self.year}"
        if self.day < 9:
            return f"0{self.day + 1}/{self.month}/{self.year}"
        else: return f"{self.day + 1}/{self.month}/{self.year}"

s = input()
day = Date(s)
if day.valid == False:
    print("INVALID")
else:
    print(day.ngayketiep())
        
            
```

```python
#W8A3
from math import sqrt

class Point:
    def __init__ (self, x, y):
        self.x = x
        self.y = y
    
    def check_goc_toado (self):
        return self.x == 0 and self.y == 0
    
    def tung (self):
        return self.x == 0
    
    def hoanh (self):
        return self.y == 0
    
    def kc (self, other):
        return f"{sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2):.2f}"

x, y = map(int, input().split())
p = Point(x, y)
if p.check_goc_toado():
    print("M la goc toa do")
elif p.tung():
    print("M nam tren truc tung")
elif p.hoanh():
    print("M nam tren truc hoanh")
else: print("M khong nam tren truc nao")

g = Point(0, 0)
print(p.kc(g))
```

```python
#W8A4
class Calculator:
    def __init__ (self, a, b):
        self.a = a
        self.b = b
    
    def add(self):
        return self.a + self.b
    
    def subtract(self):
        return self.a - self.b
    
    def product(self):
        return self.a * self.b
    
    def divide(self):
        return self.a / self.b
    
    def power(self):
        return self.a ** self.b
    
    def mod(self):
        return self.a % self.b
    
    def set_numbers(self, a, b):
        self.a = a
        self.b = b

a, b = map(int, input().split())
cal = Calculator(a, b)
while True:
    print(cal.add())
    print(cal.subtract())
    print(cal.product())
    print(cal.divide())
    print(cal.power())
    print(cal.mod())
    print("Bạn có muốn thoát?")
    s = input()
    if s == 'yes' or s == 'YES' or s == 'Yes':
        break
    else:
        x, y = map(int, input().split())
        cal.set_numbers(x, y)
```

```python
#W8A5

class ShoppingCart:
    def __init__ (self):
        self.cart = {}
        self.total = 0
    def Add (self, product, price):
        self.cart[product] = price
        self.total += price
    def Remove(self, product):
        self.total -= self.cart[product]
        del self.cart[product]
    def Check_empty(self):
        return len(self.cart) == 0
    def Total_money(self):
        return self.total
    def Hienthi(self):
        for product, price in self.cart.items():
            print(f"{product}: {price}")
    def Clear(self):
        self.cart.clear()

cart = ShoppingCart()
while True:
    q = int(input())
    if q == 2:
        product, price = map(str, input().split())
        cart.Add(product, int(price))
    elif q == 3:
        product = input()
        cart.Remove(product)
    elif q == 4:
        if cart.Check_empty(): print("Rỗng")
        else: print("Không rỗng")
    elif q == 5:
        print(cart.Total_money())
    elif q == 6:
        cart.Hienthi()
    elif q == 7:
        cart.Clear()
    
    print("Bạn có muốn dừng không?")
    s = input()
    if s == 'yes':
        break
    
```

```python
#W8A6
class Rectangle:
    def __init__ (self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return f"{self.w * self.h:.2f}"
    def perimeter(self):
        return f"{2 * (self.w + self.h):.2f}"
    def scale(self, k):
        self.w *= k
        self.h *= k

w, h, k = map(float, input().split())
rec = Rectangle(w, h)
rec.scale(k)
print(rec.area())
print(rec.perimeter())
```

```python
#W8A7
from math import gcd, lcm

def xuli(tu, mau):
    ok = 0
    if tu * mau < 0: ok = 1
    tu = abs(tu)
    mau = abs(mau)
    k = gcd(tu, mau)
    tu = tu // k
    mau = mau // k
    if ok:
        return (-tu, mau)
    return (tu, mau)

class Fraction: 
    def __init__ (self, tu, mau):
        self.tu, self.mau = xuli(tu, mau)
    
    def add(self, other):
        mauchung = lcm(self.mau, other.mau)
        tu = self.tu * (mauchung // self.mau) + other.tu * (mauchung // other.mau)
        self.tu, self.mau = xuli(tu, mauchung)

    def sub(self, other):
        mauchung = lcm(self.mau, other.mau)
        tu = self.tu * (mauchung // self.mau) - other.tu * (mauchung // other.mau)
        self.tu, self.mau = xuli(tu, mauchung)
    
    def mul(self, other):
        self.tu, self.mau = xuli(self.tu * other.tu, self.mau * other.mau)

    def div(self, other):
        self.tu, self.mau = xuli(self.tu * other.mau, self.mau * other.tu)

    def __str__(self):
        return f"{self.tu}/{self.mau}"

s = list(map(str, input().split()))
a = int(s[0])
b = int(s[1])

p1 = Fraction(a, b)
c = int(s[3])
d = int(s[4])

p2 = Fraction(c, d)
if s[2] == '+':
    p1.add(p2)
elif s[2] == '-':
    p1.sub(p2)
elif s[2] == '*':
    p1.mul(p2)
else:
    p1.div(p2)

print(p1)
    
```

```python
#W8A8
class BankAccount:
    def __init__ (self, owner, balance):
        self.owner = owner
        self.balance = balance
    def deposit (self, money):
        self.balance += money
    def withdraw(self, money):
        if self.balance >= money:
            self.balance -= money
    def __str__(self):
        return f"{self.balance:.2f}"
    
name, ib = map(str, input().split())
ib = float(ib)
owner1 = BankAccount(name, ib)
for _ in range(int(input())):
    s1, mon = map(str, input().split())
    mon = float(mon)
    if s1 == "DEPOSIT":
        owner1.deposit(mon)
    elif s1 == "WITHDRAW":
        owner1.withdraw(mon)
print(owner1)
```

```python
#W8A9
class Student:
    def __init__ (self, name):
        self.name = name
        self.total = 0
        self.cnt = 0
    def add_score(self, subject, score):
        self.total += score
        self.cnt += 1
    def avg(self):
        return f"{self.total / self.cnt:.2f}"
    def rank(self):
        t = self.total / self.cnt
        if t >= 8: return "Excellent"
        elif t < 8 and t >= 6.5: return "Good"
        elif t >= 5 and t < 6.5: return "Average"
        else: return "Poor"

name = input()
s1 = Student(name)
for _ in range(int(input())):
    s, sc = map(str, input().split())
    sc = float(sc)
    s1.add_score(s, sc)
print(f"{name} {s1.avg()} {s1.rank()}")
```

```python
#W8A10
class Book:
    def __init__ (self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
    
class Library:
    def __init__ (self):
        self.lib = []
    def add(self, book):
        self.lib.append(book)

    def count_by_author(self, name):
        cnt = 0
        for book in self.lib:
            if book.author == name: cnt += 1
        return cnt
    
    def find_by_year(self, y):
        cnt = 0
        for book in self.lib:
            if book.year == y: cnt += 1
        return cnt
    
Lib = Library()
for _ in range(int(input())):
    s1, s2 = map(str, input().split())
   
    if s1 == "COUNT":
        print(Lib.count_by_author(s2))
    elif s1 == "COUNTYEAR":
        print(Lib.find_by_year(s2))
    else:
        s = list(s2.split(";"))
        #print(s)
        book = Book(s[0], s[1], s[2])
        Lib.add(book)
```

```python
#W8A11
import math

class toado:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    def vec(self, other):
        x = other.x - self.x
        y = other.y - self.y
        z = other.z - self.z
        ans = toado(x, y, z)
        return ans
    
    def pt(self, other):
        x = self.y * other.z - self.z * other.y
        y = self.z * other.x - self.x * other.z
        z = self.x * other.y - self.y * other.x
        ans = toado(x, y, z)
        return ans
    def length(self):
        return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)
    
    def __str__(self):
        return f"{self.x} {self.y} {self.z}"

x, y, z = map(float, input().split())
a = toado(x, y, z)

x, y, z = map(float, input().split())
b = toado(x, y, z)

x, y, z = map(float, input().split())
c = toado(x, y, z)

x, y, z = map(float, input().split())
d = toado(x, y, z)

v1 = a.vec(b)
v2 = a.vec(c)
v3 = b.vec(c)
v4 = b.vec(d)

v5 = v1.pt(v2)
v6 = v3.pt(v4)

t = (v5.x * v6.x + v5.y * v6.y + v5.z * v6.z) / (v5.length() * v6.length())
t = math.acos(t)
t = math.degrees(t)
print(f"{t:.2f}")
```

```python
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = realPart
        self.imaginaryPart = imaginaryPart
    def ComplexNumbermultiply(self, other):
        return [self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart, self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart]
    
x1, x2 = map(int, input().split())
x3, x4 = map(int, input().split())
t1 = ComplexNumber(x1, x2)
t2 = ComplexNumber(x3, x4)
print(*t1.ComplexNumbermultiply(t2))
```

```python
#W8A13
from math import sqrt

def check(a, b, c):
    return (a + b > c) & (a + c > b) & (b + c > a)

def dt(a, b, c):
    p = (a + b + c) / 2
    return math.sqrt(p * (p - a) * (p - b) * (p - c))

class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def getArea(self):
        if not check(self.edge1, self.edge2, self.edge3):
            return -1
        return f"{dt(self.edge1, self.edge2, self.edge3):.2f}"
    
a, b, c = map(int, input().split())
tri = Triangle(a, b, c)
if tri.getArea() == -1: print("invalid")
else: print(tri.getArea())
```
